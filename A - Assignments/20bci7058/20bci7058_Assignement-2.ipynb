### Univariate Analysis



```python
import matplotlib.pyplot as plt
import seaborn as sns

#Histogram of a age
plt.hist(df['age'], bins=10)
plt.xlabel('age')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()

```

```python
#Bar plot of who
sns.countplot(x='who', data=df)
plt.xlabel('who')
plt.ylabel('Count')
plt.title('Bar Plot')
plt.show()
```

### Bi-Variate Analysis

```python

# Scatter plot of class and age
plt.scatter(df['pclass'], df['age'])
plt.xlabel('pclass')
plt.ylabel('age')
plt.title('Scatter Plot')
plt.show()
```

```python
# Box plot of a fare by a class
sns.boxplot(x='class', y='fare', data=df)
plt.xlabel('class')
plt.ylabel('fare')
plt.title('Box Plot')
plt.show()
```
